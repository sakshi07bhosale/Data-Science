```python
import numpy as np
# 1 to 10
arr1 = np.arange(1,11)
print("range from 1 to 10",arr1)

#zeros 3x3
zeros = np.zeros((3,3))
print("zeros 3x3\n",zeros)
# ones 2x4
ones = np.ones((2, 4))
print("ones 2x4\n",ones)
```

```python
#basic operations
import numpy as np
arr = np.array([2,4,6,8])
print("Sum:", np.sum(arr))
print("Mean:",np.mean(arr))
print("Max:",np.max(arr))
print("Min:",np.min(arr))
```

```python
#indexing and slicing
import numpy as np
arr = np.array([10,20,30,40,50])
print("First:",arr[0])
print("Last:",arr[-1])
print("Slice:",arr[1:4])
```

```python
#Element-wise Operations
import numpy as np
a = np.array([1,2,3])
b = np.array([4,5,6])
print("Addition:",a+b)
print("Multiplication:",a*b)
```

```python
#Reshape Practice
import numpy as np
arr = np.array([1,2,3,4,5,6])
print("matrix 2x3\n",arr.reshape(2,3))
print("matrix 3x2\n",arr.reshape(3,2))
```

```python
#Even Numbers
import numpy as np
arr = np.array([1,2,3,4,5,6])
even = arr[arr % 2 == 0]
print(even)
```

```python
#Replace Values (>3 → 0)
import numpy as np
arr = np.array([1,2,3,4,5,6])
arr[arr > 3] = 0
print(arr)
```

```python
#Generate Random Numbers
import numpy as np
arr = np.random.rand(5)
print(arr)
```

```python
#Matrix Operations
import numpy as np
arr = np.array([[1,2],[3,4]])
print("Total Sum:",np.sum(arr))
print("Row-wise Sum:", np.sum(arr, axis=1))
```

```python
#challeng
import numpy as np
arr = np.array([10, 20, 30, 40, 50])

# Multiply by 5
print("Multiply by 5",arr * 5)

# Elements > 25
print("Elements > 25",arr[arr > 25])

# Replace >25 with -1
arr[arr > 25] = -1
print("Replace >25 with -1",arr)
```
