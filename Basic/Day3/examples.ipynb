```python
### Task
# Show only Marks > 70
# Find average marks
# Add new column "Result" (Pass/Fail)

import pandas as pd
# Create data
data = {
    "Name": ["A", "B", "C", "D"],
    "Marks": [60, 75, 90, 50]
}

df = pd.DataFrame(data)

# 1️⃣ Show only Marks > 70
print("Marks > 70:")
print(df[df["Marks"] > 70])


# 2️⃣ Find average marks
avg = df["Marks"].mean()
print("\nAverage Marks:", avg)

# 3️⃣ Add Result column (Pass/Fail)
df["Result"] = df["Marks"].apply(lambda x: "Pass" if x >= 60 else "Fail")

print("\nUpdated DataFrame:")
print(df)
```
