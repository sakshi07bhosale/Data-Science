```python
#create data
import pandas as pd
import numpy as np

data = {
    "Name": ["A", "B", "C", "D"],
    "Marks": [80, np.nan, 90, None],
    "Age": [20, 21, None, 22]
}

df = pd.DataFrame(data)
print(df)
```

```python
#Check Missing Values
print(df.isnull())        # shows True/False
```

```python
#missing value
print(df.isnull().sum())  # count missing
```
