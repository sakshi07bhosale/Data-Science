```python
#Load Data
import pandas as pd

df = pd.read_csv("Student_dataset.csv")
print(df.head())
```

```python
#Handle Missing Values
df["Marks"].fillna(df["Marks"].mean(),inplace=True)
```

```python
df["Age"].fillna(df["Age"].median(),inplace=True)
```

```python
df["City"].fillna(df["City"].mode()[0],inplace=True)
```

```python
#Create New Columns
df["Result"]=df["Marks"].apply(lambda x: "Pass" if x >= 75 else "Fail")
print(df)
```

```python
#create new Grade Column
def grade(x):
    if x >= 85:
        return "A"
    elif x >=70:
        return "B"
    else:
        return "C"
df["Grade"]=df["Marks"].apply(grade)
print(df)
```

```python
#Scaling (0 to 1)
df["Marks_scaled"] = (df["Marks"] - df["Marks"].min()) / (df["Marks"].max() - df["Marks"].min())
print(df)
```

```python
#Encoding (City → Numbers)
df_encoded = pd.get_dummies(df["City"])
print(df_encoded.head())
```
