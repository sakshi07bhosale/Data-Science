```python
#Basic Feature Creation
#result column
import pandas as pd 
df = pd.read_csv("Student_dataset.csv")
df["Result"] = df["Marks"].apply(lambda x: "Pass" if x >= 75 else "Fail")
print(df)
```

```python
#add gradde column
def grade(x):
    if x >= 85:
        return "A"
    elif x >= 70:
        return "B"
    else:
        return "C"
df["Grade"] = df["Marks"].apply(grade)
print(df)
```

```python
#Binning (Grouping Values)
df["Marks_Category"] = pd.cut(df["Marks"],
                             bins=[0, 60, 80, 100],
                             labels=["Low", "Medium", "High"])
print(df)
```

```python
#Label Encoding
df["City_Code"] = df["City"].astype("category").cat.codes
print(df)
```

```python
#One-Hot Encoding
df = pd.get_dummies(df, columns=["City"])
print(df)
```

```python
#Scaling
df["Marks_scaled"] = (df["Marks"] - df["Marks"].min()) / (df["Marks"].max() - df["Marks"].min())
print(df)
```
