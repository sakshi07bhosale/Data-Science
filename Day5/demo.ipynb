```python
#Creating New Columns
import pandas as pd

data = {
    "Name": ["A", "B", "C"],
    "Marks": [80, 60, 90]
}

df = pd.DataFrame(data)

# Add new column
df["Result"] = df["Marks"].apply(lambda x: "Pass" if x >= 70 else "Fail")

print(df)
```

```python
#Mapping
import pandas as pd
data = {
    "Grade": ["A", "B", "A", "C"]
}

df = pd.DataFrame(data)

df["Grade"] = df["Grade"].map({"A": 3, "B": 2, "C": 1})

print(df)
```

```python
#Scaling Data
data = {
    "Marks": [50, 60, 70, 80]
}

df = pd.DataFrame(data)

df["Marks_scaled"] = (df["Marks"] - df["Marks"].min()) / (df["Marks"].max() - df["Marks"].min())

print(df)
```
