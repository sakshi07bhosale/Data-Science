```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create data (with extra student for practice)
data = {
    "Name": ["A", "B", "C", "D", "E"],
    "Marks": [85, 70, 90, 60, 75]
}

df = pd.DataFrame(data)

# 1️⃣ Bar Chart
plt.figure()
plt.bar(df["Name"], df["Marks"])
plt.title("Student Marks - Bar Chart")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()
```
