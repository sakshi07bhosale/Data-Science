```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create data (with extra student for practice)
data = {
    "Name": ["A", "B", "C", "D", "E"],
    "Marks": [85, 70, 90, 60, 75]
}

df = pd.DataFrame(data)
```

```python

# 1️⃣ Bar Chart
plt.figure()
plt.bar(df["Name"], df["Marks"])
plt.title("Student Marks - Bar Chart")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()
```

```python

# 2️⃣ Line Chart
plt.figure()
plt.plot(df["Name"], df["Marks"], marker='o')
plt.title("Marks Trend - Line Chart")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()
```

```python
# 3️⃣ Pie Chart
plt.figure()
plt.pie(df["Marks"], labels=df["Name"], autopct="%1.1f%%")
plt.title("Marks Distribution - Pie Chart")
plt.show()
```
