```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dataset
data = {
    "Name": ["A", "B", "C", "D", "E"],
    "Marks": [85, 70, 90, 60, 75],
    "Age": [20, 21, 19, 22, 20]
}

df = pd.DataFrame(data)

# Create multiple plots (dashboard style)
plt.figure(figsize=(10, 6))
```

```python
# 1. Bar chart
plt.subplot(2, 2, 1)
sns.barplot(x="Name", y="Marks", data=df)
plt.title("Marks")
```

```python
# 2. Histogram
plt.subplot(2, 2, 2)
plt.hist(df["Marks"])
plt.title("Marks Distribution")
```

```python
# 3. Boxplot
plt.subplot(2, 2, 3)
sns.boxplot(y=df["Marks"])
plt.title("Boxplot")
```

```python
# 4. Line chart
plt.subplot(2, 2, 4)
plt.plot(df["Marks"], marker='o')
plt.title("Trend")
```
