```python
#Add Column
df["Result"] = df["Marks"].apply(lambda x: "Pass" if x >= 50 else "Fail")
print(df)
```

```python
#Data Cleaning
df.dropna()  #remove missing
```

```python
df.fillna(0)  #replace missing
```

```python
df.drop_duplicates()  #remove duplicates
```

```python
#Data Transformation
df["Marks"] = df["Marks"] * 2
df.rename(columns={"Marks": "Score"}, inplace=True)
print(df)
```
