```python
import pandas as pd
import numpy as np

data = {
    "Name": ["A", "B", "C", "D", "E", "F"],
    "Marks": [85, 70, None, 60, 90, None],
    "Age": [20, 21, 19, None, 22, 20],
    "City": ["Mumbai", "Pune", "Mumbai", "Delhi", "Pune", "Mumbai"]
}

df = pd.DataFrame(data)
print(df)
```

```python
# 1️⃣ Handle Missing Values
df["Marks"].fillna(df["Marks"].mean(),inplace=True)
df["Age"].fillna(df["Age"].median(),inplace=True)
```

```python
# 2️⃣ Add Result Column
df["Result"] = df["Marks"].apply(lambda x: "Pass" if x >= 75 else "Fail")
print(df)
```
