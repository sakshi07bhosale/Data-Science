```python
import pandas as pd

# Create data
data = {
    "Name": ["A", "B", "C", "D"],
    "Marks": [60, 75, 90, 50],
    "City": ["Mumbai", "Pune", "Mumbai", "Delhi"]
}

df = pd.DataFrame(data)
print(df)
```

```python
# 1️⃣ Students with Marks > 70
print("Marks > 70:\n", df[df["Marks"] > 70])
```

```python
# 2️⃣ Average Marks
print("\nAverage Marks:", df["Marks"].mean())
```

```python
# 3️⃣ Group by City (Average Marks)
print("\nAverage marks by city:\n", df.groupby("City")["Marks"].mean())
```
