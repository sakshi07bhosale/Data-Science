```python
import pandas as pd
import numpy as np

data = {
    "Name": ["A", "B", "C", "D", "E", "F"],
    "Marks": [85, 70, None, 60, 90, None],
    "Age": [20, 21, 19, None, 22, 20],
    "City": ["Mumbai", "Pune", "Mumbai", "Delhi", "Pune", "Mumbai"]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Fill missing values
df["Marks"].fillna(df["Marks"].mean(), inplace=True)
df["Age"].fillna(df["Age"].median(), inplace=True)
```

```python
# Add Result column
df["Result"] = df["Marks"].apply(lambda x: "Pass" if x >= 75 else "Fail")
print(df)
```

```python
#Analysis
print("Average Marks:", df["Marks"].mean())
print("\nMarks by City:\n", df.groupby("City")["Marks"].mean())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1️⃣ Bar Chart (Marks by Student)
plt.figure()
sns.barplot(x="Name", y="Marks", data=df)
plt.title("Marks by Student")
plt.show()
```
