```python
import pandas as pd

data = {
    "Marks": [50, 60, 70, 80, 90]
}

df = pd.DataFrame(data)
print(df)
```

```python
#mean
print(df["Marks"].mean())
```

```python
#median
print(df["Marks"].median())
```

```python
#mode
print(df["Marks"].mode())
```

```python
#find min nd max
print(df["Marks"].min())
print(df["Marks"].max())
```

```python
#range
print(df["Marks"].max() - df["Marks"].min())
```

```python
#Variance
print(df["Marks"].var())
```

```python
#Standard Deviation
print(df["Marks"].std())
```

```python
print(df.describe())
```
