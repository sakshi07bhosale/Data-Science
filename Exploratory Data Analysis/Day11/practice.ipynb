```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv("Student_dataset.csv")
```

```python
df["Marks"].fillna(df["Marks"].mean(), inplace=True)
df["Age"].fillna(df["Age"].median(), inplace=True)
df["City"].fillna(df["City"].mode()[0], inplace=True)
```

```python
#Bar Chart (Marks by City)

city_avg = df.groupby("City")["Marks"].mean()

plt.bar(city_avg.index, city_avg.values)
plt.title("Average Marks by City")
plt.xlabel("City")
plt.ylabel("Marks")
plt.show()
```

```python
#Line Chart (Marks Trend)
plt.plot(df["Marks"])
plt.title("Marks trend")
plt.xlabel("Index")
plt.ylabel("Marks")
plt.show()
```

```python
#Histogram (Distribution)
plt.hist(df["Marks"])
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
#Box Plot (Outliers Detection)
sns.boxplot(y=df["Marks"])
plt.title("Marks Box Plot")
plt.show()
```

```python
#Seaborn Barplot
sns.barplot(x="City",y="Marks",data=df)
plt.title("Marks by City (Seaborn)")
plt.show()
```

```python
#Heatmap

corr =df.corr(numeric_only=True)

sns.heatmap(corr,annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
