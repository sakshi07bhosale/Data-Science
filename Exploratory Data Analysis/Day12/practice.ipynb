```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("Student_dataset.csv")
```

```python
# Clean
df["Marks"].fillna(df["Marks"].mean(), inplace=True)
df["Age"].fillna(df["Age"].median(), inplace=True)
df["City"].fillna(df["City"].mode()[0], inplace=True)
```

```python
#Pattern: Which City Performs Better?
city_avg = df.groupby("City")["Marks"].mean()
print(city_avg)

sns.barplot(x=city_avg.index,y=city_avg.values)
plt.title("Average Marks by city")
plt.show()
```

```python
#Pattern: Relationship (Age vs Marks)
sns.scatterplot(x="Age", y="Marks", data=df)
plt.title("Age vs Marks")
plt.show()
```

```python
#Pattern: Distribution of Marks

plt.hist(df["Marks"])
plt.title("Marks Distribution")
plt.show()
```

```python
#Pattern: Outliers
sns.boxplot(y=df["Marks"])
plt.title("Marks Boxplot")
plt.show()
```
