```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.DataFrame({
    "Hours": [1,2,3,4,5],
    "Marks": [20,40,50,70,90]
})

print(df)
```

```python
#Input & Output

X = df[["Hours"]]
y = df["Marks"]
```

```python
#Create Model

model = LinearRegression()
```

```python
#Train Model

model.fit(X, y)
```

```python
#Predict

pred = model.predict([[6]])

print(pred)
```

```python
##Visualization


import matplotlib.pyplot as plt

# Scatter plot
plt.scatter(
    df["Hours"],
    df["Marks"]
)

# Regression line
plt.plot(
    df["Hours"],
    model.predict(X)
)

plt.xlabel("Hours")
plt.ylabel("Marks")

plt.title("Linear Regression")

plt.show()
```
