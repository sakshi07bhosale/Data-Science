```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
#Create Dataset

df = pd.DataFrame({
    "Hours": [1,2,3,4,5,6],
    "Result": [0,0,0,1,1,1]
})

print(df)
```

```python
#Input & Output

X = df[["Hours"]]
y = df["Result"]
```

```python
#Create Model

model = DecisionTreeClassifier()
```

```python
#Train Model

model.fit(X, y)
```

```python
#Predict

pred = model.predict([[5]])

print(pred)
```

```python
#Visualize Decision Tree

from sklearn import tree
import matplotlib.pyplot as plt

tree.plot_tree(model)

plt.show()
```
