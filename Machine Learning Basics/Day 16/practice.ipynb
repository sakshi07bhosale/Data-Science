```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# Data
df = pd.DataFrame({
    "Hours": [1, 2, 3, 4, 5],
    "Marks": [20, 40, 50, 70, 90]
})
```

```python

# Input and output
X = df[["Hours"]]
y = df["Marks"]
```

```python
#model
model=LinearRegression()
```

```python
# Train
model.fit(X, y)
```

```python

# Predict
pred = model.predict([[6]])
```

```python
print("Predicted Marks:", pred[0])
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["Hours"], df["Marks"])
plt.plot(df["Hours"], model.predict(X))
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.title("Regression Line")
plt.show()
```
