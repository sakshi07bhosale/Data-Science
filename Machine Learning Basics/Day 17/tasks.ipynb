```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
# Data
df = pd.DataFrame({
    "Hours": [1, 2, 3, 8, 9, 10],
    "Marks": [20, 25, 30, 80, 85, 90]
})
```

```python

# Features
X = df[["Hours", "Marks"]]
X
```

```python
# Model
model = KMeans(n_clusters=2, random_state=42)
```

```python
# Train
model.fit(X)

# Add cluster column
df["Cluster"] = model.labels_
```

```python

print(df)
```

```python
#Create 3 Clusters
model = KMeans(n_clusters=3, random_state=42)

model.fit(X)

df["Cluster"] = model.labels_

print(df)
```

```python
#Customer Segmentation
df = pd.DataFrame({
    "Income": [20,25,30,80,85,90],
    "Spending": [30,35,40,85,90,95]
})

X = df[["Income", "Spending"]]

model = KMeans(n_clusters=2, random_state=42)

model.fit(X)

df["Cluster"] = model.labels_

print(df)
```

```python
#Visualization

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
plt.scatter(
    df["Income"],
    df["Spending"],
    c=df["Cluster"]
)

plt.xlabel("Income")
plt.ylabel("Spending")

plt.title("Customer Segments")

plt.show()
```

```python
#Elbow Method
wcss = []

for i in range(1, 6):

    model = KMeans(
        n_clusters=i,
        random_state=42
    )

    model.fit(X)

    wcss.append(model.inertia_)

print(wcss)
```

```python
#Plot Elbow Graph
plt.plot(range(1, 6), wcss)

plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")

plt.title("Elbow Method")

plt.show()
```
