```python
from sklearn.metrics import accuracy_score

y_true = [1,0,1,1,0]
y_pred = [1,0,1,0,0]

acc = accuracy_score(y_true, y_pred)

print(acc)
```

```python
#precision
from sklearn.metrics import precision_score

precision = precision_score(
    y_true,
    y_pred
)

print(precision)
```

```python
#recall
from sklearn.metrics import recall_score

recall = recall_score(
    y_true,
    y_pred
)

print(recall)
```

```python

#Confusion Matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(
    y_true,
    y_pred
)

print(cm)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

import pandas as pd

# Data
df = pd.DataFrame({
    "Hours": [1,2,3,4,5,6],
    "Result": [0,0,0,1,1,1]
})

# Input/output
X = df[["Hours"]]
y = df["Result"]

# Model
model = LogisticRegression()

# Train
model.fit(X, y)

# Predict
y_pred = model.predict(X)

# Metrics
print("Accuracy:",
      accuracy_score(y, y_pred))

print("Precision:",
      precision_score(y, y_pred))

print("Recall:",
      recall_score(y, y_pred))
```
