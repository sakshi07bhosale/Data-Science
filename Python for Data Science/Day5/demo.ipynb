```python
#Creating New Columns
import pandas as pd

data = {
    "Name": ["A", "B", "C"],
    "Marks": [80, 60, 90]
}

df = pd.DataFrame(data)

# Add new column
df["Result"] = df["Marks"].apply(lambda x: "Pass" if x >= 70 else "Fail")

print(df)
```

```python
import pandas as pd

# Create data
data = {
    "Name": ["A", "B", "C"],
    "Marks": [45, 75, 85],
    "Grade": ["C", "B", "A"],
    "City": ["Mumbai", "Pune", "Mumbai"]
}

df = pd.DataFrame(data)

# 1. Add Result column
df["Result"] = df["Marks"].apply(lambda x: "Pass" if x >= 50 else "Fail")

# 2. Mapping (Grade → number)
df["Grade"] = df["Grade"].map({"A": 3, "B": 2, "C": 1})

# 3. Scaling Marks (0 to 1)
df["Marks_scaled"] = (df["Marks"] - df["Marks"].min()) / (df["Marks"].max() - df["Marks"].min())

# 4. Change data type
df["Marks"] = df["Marks"].astype(float)

# 5. Add Bonus (+5)
df["Bonus"] = df["Marks"].apply(lambda x: x + 5)

# 6. Encoding (City)
city_encoded = pd.get_dummies(df["City"])

# 7. Rename column
df.rename(columns={"Marks": "Score"}, inplace=True)

# 8. Combine columns
df["Info"] = df["Name"] + " - " + df["Result"]

# Output
print(df)
print("\nEncoded City Data:\n", city_encoded)
```
