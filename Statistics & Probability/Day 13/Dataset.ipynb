```python
import pandas as pd 
df =  pd.read_csv("Student_dataset.csv")
```

```python
#mean
df["Marks"].fillna(df["Marks"].mean(),inplace=True)
```

```python
#mean
mean=df["Marks"].mean()
print(mean)
```

```python
#median
median=df["Marks"].median()
print(median)
```

```python
#mode
mode = df["Marks"].mode()[0]
print(mode)
```

```python
#varience
varaience=df["Marks"].var()
print(varaience)
```

```python
import matplotlib.pyplot as plt

plt.hist(df["Marks"])
plt.axvline(mean, linestyle="--")
plt.axvline(median, linestyle="--")
plt.title("Marks Distribution")
plt.show()
```
