```python
import pandas as pd 
data = {
    "Marks":[50,60,70,80,90]
}
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
#mean
print(df["Marks"].mean())
```

```python
#median
print(df["Marks"].median())
```

```python
#mode
print(df["Marks"].mode())
```

```python
#varience
print(df["Marks"].var())
```
