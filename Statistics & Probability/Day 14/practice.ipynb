```python
import pandas as pd

df = pd.read_csv("Student_dataset.csv")
df["Marks"].fillna(df["Marks"].mean(), inplace=True)

prob = len(df[df["Marks"] < 60]) / len(df)
print(prob)
```

```python
prob = len(df[df["City"] == "Mumbai"]) / len(df)
print(prob)
```

```python
pune = df[df["City"] == "Pune"]

prob = len(pune[pune["Marks"] > 75]) / len(pune)
print(prob)
```

```python
prob = len(df[(df["Marks"] > 75) & (df["Age"] > 21)]) / len(df)
print(prob)
```

```python
prob = len(df[(df["Marks"] > 75) & (df["City"] == "Delhi")]) / len(df)
print(prob)
```
